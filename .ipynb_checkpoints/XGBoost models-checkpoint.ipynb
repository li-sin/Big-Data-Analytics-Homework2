## Define a function for modeling and cross-validation

This function will do the following:
1. fit the model
2. determine training accuracy
3. update n_estimators with cv function of xgboost package
4. plot Feature Importance 

#### 更改的地方：

1. 用auc測量Multiclass的資料，會有error，
所以我將auc相關的Code都註解掉，
而cv和fit的參數metrics、eval_metric我都改成merror，
merror可以用在Multiclass的資料上

2. 而且也加上參數
xgb_param['num_class'] = 10，
xgboost可以跑Multiclass的資料

3. 

```python
#test_results = pd.read_csv('test_results.csv')
def modelfit(alg, dtrain, dtest, predictors,useTrainCV=True, cv_folds=10, early_stopping_rounds=50):
    
    if useTrainCV:
        xgb_param = alg.get_xgb_params()
        xgb_param['num_class'] = 10
        #Set number of label
        xgtrain = xgb.DMatrix(dtrain[predictors].values, label=dtrain[target].values)
        xgtest = xgb.DMatrix(dtest[predictors].values)
        cvresult = xgb.cv(xgb_param, xgtrain, num_boost_round=alg.get_params()['n_estimators'], nfold=cv_folds,
            metrics='merror', early_stopping_rounds=early_stopping_rounds)#, show_progress=False)
        alg.set_params(n_estimators=cvresult.shape[0])
    
    #Fit the algorithm on the data
    #alg.fit(dtrain[predictors], dtrain['Class'],eval_metric='auc')
    alg.fit(dtrain[predictors], dtrain['Class'],eval_metric='merror')
        
    #Predict training set:
    dtrain_predictions = alg.predict(dtrain[predictors])
    dtrain_predprob = alg.predict_proba(dtrain[predictors])[:,1]
        
    #Print model report:
    print "\nModel Report"
    print "Accuracy : %.4g" % metrics.accuracy_score(dtrain['Class'].values, dtrain_predictions)
    #print "Coverage_error( : %.4g" % metrics.coverage_error(dtrain['Class'].values, dtrain_predictions)
    #print "AUC Score (Train): %f" % metrics.roc_auc_score(dtrain['Class'], dtrain_predprob)
    
#     Predict on testing data:
    #dtest['predprob'] = alg.predict_proba(dtest[predictors])[:,1]
    #results = test_results.merge(dtest[['ID','predprob']], on='ID')
    #print 'AUC Score (Test): %f' % metrics.roc_auc_score(results['Class'], results['predprob'])
     
    print cvresult
    
    feat_imp = pd.Series(alg.booster().get_fscore()).sort_values(ascending=False)
    feat_imp.plot(kind='bar', title='Feature Importances')
    plt.ylabel('Feature Importance Score')
    return feat_imp
```

# 這裡將modelfit做修改，讓它回傳cvresult整個dataframe而不是只是印出來


```python
#test_results = pd.read_csv('test_results.csv')
def modelfit2(alg, dtrain, dtest, predictors,useTrainCV=True, cv_folds=10, early_stopping_rounds=50):
    
    if useTrainCV:
        xgb_param = alg.get_xgb_params()
        xgb_param['num_class'] = 10
        #Set number of label
        xgtrain = xgb.DMatrix(dtrain[predictors].values, label=dtrain[target].values)
        xgtest = xgb.DMatrix(dtest[predictors].values)
        cvresult  = xgb.cv(xgb_param, xgtrain, num_boost_round=alg.get_params()['n_estimators'], nfold=cv_folds,
            metrics='merror', early_stopping_rounds=early_stopping_rounds)#, show_progress=False)
        alg.set_params(n_estimators=cvresult.shape[0])
    
    #Fit the algorithm on the data
    #alg.fit(dtrain[predictors], dtrain['Class'],eval_metric='auc')
    alg.fit(dtrain[predictors], dtrain['Class'],eval_metric='merror')
        
    #Predict training set:
    dtrain_predictions = alg.predict(dtrain[predictors])
    dtrain_predprob = alg.predict_proba(dtrain[predictors])[:,1]
        
    #Print model report:
    print "\nModel Report"
    print "Accuracy : %.4g" % metrics.accuracy_score(dtrain['Class'].values, dtrain_predictions)
    #print "Coverage_error( : %.4g" % metrics.coverage_error(dtrain['Class'].values, dtrain_predictions)
    #print "AUC Score (Train): %f" % metrics.roc_auc_score(dtrain['Class'], dtrain_predprob)
    
#     Predict on testing data:
    #dtest['predprob'] = alg.predict_proba(dtest[predictors])[:,1]
    #results = test_results.merge(dtest[['ID','predprob']], on='ID')
    #print 'AUC Score (Test): %f' % metrics.roc_auc_score(results['Class'], results['predprob'])
     
    print cvresult
    
    feat_imp = pd.Series(alg.booster().get_fscore()).sort_values(ascending=False)
    feat_imp.plot(kind='bar', title='Feature Importances')
    plt.ylabel('Feature Importance Score')
    return feat_imp,cvresult
```

```python
xgb4 = XGBClassifier(
        learning_rate =0.05,
        n_estimators=280,
        max_depth=5,
        min_child_weight=3,
        gamma=0,
        subsample=0.7,
        colsample_bytree=0.6,
        reg_alpha=1e-7,
        objective= 'multi:softmax',
        nthread=4,
        scale_pos_weight=1,
        seed=27)
feat_imp4,cvresult4 = modelfit2(xgb4, train, test, predictors)

```
